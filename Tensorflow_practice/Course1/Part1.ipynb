## 1. 导入包

```python
import tensorflow as tf
import numpy as np
from tensorflow import keras
```

## 2. 定义和编译神经网络
创建一个很简单的神经网络，它只有一层，一个神经元，并且输入的也是一个值

```python
# 实例化一个序列化模型
model = tf.keras.Sequential([keras.layers.Dense(units = 1, input_shape = [1])])
```

接下来，让我们编译这个神经网络，但是我们必须要明确两个函数：损失函数和优化函数

我们人类可以很轻松的得知这些数之间的关系是y=2x-1，但是计算机没那么简单。

**loss：**当计算机尝试“学习”时，会做出猜测……也许是y = 10x + 10。 损失函数可以根据已知的正确答案来衡量猜测的答案，并衡量其执行的好坏程度。

**optimizer：**然后使用优化器函数进行另一次猜测。根据损失函数的运行情况，它会尽量减少损失。它可能会得到像y=5x+5这样的结果，虽然仍然很糟糕，但更接近正确的结果(即损失更低)

神经网络将重复此操作，我们让它使用“均方误差”来表示损失，并使用“随机梯度下降”来表示优化器的方法。 (您还不需要了解它们的数学原理，但是您可以看到它们起作用了！)

```python
model.compile(optimizer='sgd',loss='mean_squared_error')
```

随着不断学习，将了解针对不同情况的不同且适当的损失和优化器功能。

## 3. 导入数据
接下来，将输入一些数据给神经网络。 

```python
xs = np.array([-1.0,  0.0, 1.0, 2.0, 3.0, 4.0], dtype=float)
ys = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0], dtype=float)
```

## 4. 训练神经网络
训练神经网络的过程是在model.fit调用中“学习” Xs和Ys之间的关系。 这就是我们上面讨论过的循环的地方，将对关系进行猜测，测量其好坏（也称为损失），使用优化器进行另一次猜测等。它将针对您所指定的循环周期数进行运行。运行此代码时，可以在右侧看到损失。

```python
model.fit(xs,ys,epochs=500)
```

现在有了一个经过训练的模型来学习X和y之间的关系，你可以使用这个模型。来预测Y的值

例如：如果X = 10，你认为Y会是多少?

```python
print(model.predict([10.0]))
```

**Question：**
 * 你也许会认为输出的是19，但是结果比这小一点，是18.982117，为什么会这样呢？

**Answer：**
 * 请记住，神经网络对概率进行预测，因此，提供给神经网络的数据，它计算出X和Y之间的关系为Y = 2X-1的可能性非常高，但是只有6个数据点，我们不能确定Y=2X-1一定成立，所以预测的结果非常接近19，但不一定是19。
 
当您使用神经网络时，您将看到这种模式重复出现。你几乎总是处理概率,不确定性,并将做一些编码算出结果是基于概率的,特别是在分类过程中。
