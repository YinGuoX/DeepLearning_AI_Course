# 使用Tokenizer处理文本数据

生成对应的词字典

```
from tensorflow.keras.preprocessing.text import Tokenizer

sentences = [
             'i love my dog',
             'I, love my cat',
             'You love my dog!'
]

# num_words是处理的最大单词数量
tokenizer = Tokenizer(num_words = 100)
# 实现分词：生成token词典
tokenizer.fit_on_texts(sentences)

# 词与索引的对应关系
word_index = tokenizer.word_index

print(word_index)
```
